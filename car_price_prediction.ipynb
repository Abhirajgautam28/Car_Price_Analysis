#### Line plot

```python
sns.lineplot(x = "Brand", y = "Price" , data = car_1)
plt.title("Average Car price by Brand")
plt.xlabel("Cars Brands")
plt.ylabel("Prices of different cars according to Brand")
plt.tight_layout()
plt.xticks(rotation = (45))
plt.show()
```

```python
sns.lineplot(x = "Fuel Type", y = "Price" , data = car_1)
plt.title("Average Car price by its Fuel Type")
plt.xlabel("Cars Fuel type")
plt.ylabel("Prices of different cars according to Fuel type")
plt.tight_layout()
plt.xticks(rotation = (45))
plt.show()
```

```python
sns.lineplot(x = "Condition", y = "Price" , data = car_1)
plt.title("Average Car price by its Condition")
plt.xlabel("Cars Condition")
plt.ylabel("Prices of different cars according to their Condition")
plt.tight_layout()
plt.xticks(rotation = (45))
plt.show()
```

```python
sns.lineplot(x = "Year", y = "Price" , data = car_1)
plt.title("Average Car price by the Year")
plt.xlabel("Years of cars")
plt.ylabel("Prices of different cars according to the Years")
plt.tight_layout()
plt.xticks(rotation = (45))
plt.show()
```

#### Count plot

```python
sns.countplot(x = "Transmission",data =car_1)
plt.title("Comparison between Transmission of different cars")
plt.xlabel("Transmission of cars")
plt.grid(True,alpha = 0.8)
plt.show()
```

#### Pie Chart

```python
plt.figure(figsize = (8,6))
car = car_1["Brand"].value_counts()
plt.pie(car, labels = car.index , autopct = '%1.1f%%', startangle = 90)
plt.title("Brand-wise Car Count in Dataset")
plt.axis("equal")
plt.tight_layout()
plt.grid(True , alpha = 0.9999)
plt.show()
```

#### Bar Plot

```python
plt.figure(figsize = (8,6))
v_1 = car_1["Model"].value_counts()
v_1.plot(kind = 'bar', color = 'steelblue')
plt.title("Most Common Car Models in Dataset")
plt.xlabel("Car Model")
plt.ylabel("Number of Cars")
plt.tight_layout() 
plt.grid(True , alpha =0.9)
plt.show()
```
